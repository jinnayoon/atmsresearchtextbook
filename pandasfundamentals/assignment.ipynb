7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger

```python
mag_4_higher = df[df["mag"] >= 4]
mag_4_higher
```

8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes. Location name on the x axis, Earthquake count on the y axis

```python
unique_mag4_countries = mag_4_higher["country"].unique()
#iterate thru and if it pops up add it? how do i do this efficiently instead of hardcoding
#do i use a function count() that adds up the condition that country = "country" 
#but how do i sort without groupby?
counts = {} #make empty dictionary to store the location/country and their values
#the reason why we need a dictionary and not a list is because there are more features/values we have to consider of the entire row
#prob should have made this into a dataframe
for location in unique_mag4_countries:
    temp_df = mag_4_higher[mag_4_higher["country"] == location]
    counts[location] = len(temp_df)
counts = pd.Series(counts)
#largest 
largest_5 = counts.nlargest(5)
largest_5
```

```python
from matplotlib import pyplot as plt
#we can use iloc for series
barchart = largest_5.plot.bar()
```

9) Make a histogram the distribution of the Earthquake magnitudes. Do one subplot for the filtered and one for the unfiltered dataset. Use a Logarithmic scale. What sort of relationship do you see?

```python
fig, ax = plt.subplots(2, 1, figsize = (8,5))
#unflitered
ax[0].hist(df["mag"])
ax[0].set_yscale("log") #using logarithmic scale to make everything the same comparison for y axis
ax[0].set_xlabel("Earthquake Magnitude")
ax[0].set_ylabel("Frequency")
#filtered
ax[1].hist(mag_4_higher["mag"])
ax[1].set_yscale("log")
ax[1].set_xlabel("Earthquake Magnitude")
ax[1].set_ylabel("Frequency")
```

There is more of a normal distribution between the unfiltered dataset while the filtered one has a peak frequency at 4-5 and then has less as the magnitudes get higher. You can see more specific values in the filtered histogram. In the middle ranges of the unfiltered histogram, you can see that ranges 1-4 contain the most occurences.

11) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude. Use a two-column subplot with both the filtered and unfiltered datasets. Color the points by magnitude. Make it pretty. What difference do you note between the filtered and unfiltered datasets?

```python
fig, ax = plt.subplots(ncols = 2, figsize = (20,5))
scat1 = ax[0].scatter(df["longitude"], df["latitude"], s = 5, c = df["mag"]) #s is size which makes it smaller
scat2 = ax[1].scatter(mag_4_higher["longitude"], mag_4_higher["latitude"], s = 5, c = mag_4_higher["mag"])
ax[0].set_xlabel("Longitude")
ax[0].set_ylabel("Latitude")
ax[1].set_xlabel("Longitude")
ax[1].set_ylabel("Latitude")
ax[0].set_title("Unfiltered")
ax[1].set_title("Filtered")
#include color ba/key, need the range to be the magnitude 
fig.colorbar(scat1)
fig.colorbar(scat2)
```

I noticed that there are more occurences of earthquakes in magnitude between 2-6 than there is of the filtered one that has more occurences between 4-6. This makes sense as seen with the observations from the histogram. The distribution of earthquakes in the filtered scatter plot also shows that most of the earthquakes are concentrated on the edge of the countries- where the tectonic plates are. We can see that most of the earthquake clusters in both filtered and unfiltered scatter plots are around South Asia and South America. These countries must experience many earthquakes due to the geographical position (and other factors) both in low and high magnitude.
